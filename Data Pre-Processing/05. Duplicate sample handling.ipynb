```python
import pandas as pd
df = pd.read_csv('../Datasets/diabetes_two.csv')
```

<h1><b><font color="lime">Handling Duplicate Samples</b></h1>

```python
#finding the duplicates
duplicates = df[df.duplicated(keep='first')]
duplicates
```

<font color = "yellow">
(i) dropping duplicate samples

```python
new_df = df.drop_duplicates()
```

```python
#rechecking duplicates
duplicates = new_df[new_df.duplicated(keep='first')]
duplicates
```

```python
print('previous df size: ', len(df))
print('new df size: ', len(new_df))
```
