```python
import pandas as pd
df = pd.read_csv('../Datasets/diabetes_two.csv')
```

<h1><b><font color="green">
Handling Missing Values

<font color="yellow">
(i) Dropping

```python
#check every cell for null values
df.isnull()
```

```python
#count of null values for each column
df.isnull().sum()
```

```python
#dataset information. null count and datatype for each column
df.info()
```

```python
#numerical data information
df.describe()
```

```python
#creating a copy of df
df1 = df.copy()
```

```python
#dropping the rows containing null cells
df1.dropna(subset=['Age'], inplace=True)
df1.dropna(subset=['Gender'], inplace=True)
df1.dropna(subset=['weakness'], inplace=True)
#inplace=True makes sure that the modification takes place directly and nothing is returned
```

```python
#check dataset length changes
print('old size: ',len(df))
print('new size: ',len(df1))
```

<font color="yellow">
(ii) fill with mean/median/most frequent value

```python
df2 = df.copy()
```

```python
#calculate mean value for Age column
ageMean = df2['Age'].mean()
#replace null cells with mean value
df2['Age'] = df2['Age'].fillna(ageMean)
```

```python
#replace with median value
ageMedian = df2['Age'].median()
df2['Age'] = df2['Age'].fillna(ageMedian)
```

```python
#replace with most frequent value (Gender column)
most_freq = df2['Gender'].mode().iloc[0]
df2['Gender'] = df2['Gender'].fillna(most_freq)
```

```python
#replace with most frequent value (weakness column)
most_freq = df2['weakness'].mode().iloc[0]
df2['weakness'] = df2['weakness'].fillna(most_freq)
```

```python
#check null counts for each column
df2.isnull().sum()
```

```python
#fill with most freq. value with lambda programming
# same thing but with lambda programming
df2 = df2.apply(lambda x:x.fillna(x.value_counts().index[0]))
```
